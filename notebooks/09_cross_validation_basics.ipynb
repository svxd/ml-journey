```python
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("../data/raw/titanic.csv")
df.head()
```

```python
features = ["Pclass", "Sex", "Age", "Fare", "Embarked"]
target = "Survived"

X = df[features]
y = df[target]

X.head()
```

```python
numeric_features = ["Pclass", "Age", "Fare"]
categorical_features = ["Sex", "Embarked"]
```

```python
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000))
])
```

```python
scores = cross_val_score(
    model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

scores
```

```python
scores.mean(), scores.std()
```

```python
# Task A
features = ["Pclass", "Sex", "Age", "Fare", "Embarked"]
target = "Survived"

X = df[features]
y = df[target]

numeric_features = ["Pclass", "Age", "Fare"]
categorical_features = ["Sex", "Embarked"]
```

```python
# Task B
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
# Task C
model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000))
])
```

```python
# Task D
scores = cross_val_score(
    model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

scores
```

```python
# Task E

print(f"CV mean accuracy: {scores.mean():.4f}")
print(f"CV std accuracy: {scores.std():.4f}") 
```

## Выводы
- Один train/test split может дать случайно удачную или неудачную оценку
- Cross-validation даёт более стабильную оценку качества модели
- Pipeline внутри CV помогает избежать data leakage
