```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/employees_groupby.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df["department"].value_counts().plot(kind="bar")
plt.title("Employees per Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
df["salary"].plot(kind="hist", bins=5)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.show()
```

```python
df["salary"].plot(kind="box")
plt.title("Salary Boxplot")
plt.show()
```

```python
df.groupby("department")["salary"].mean().plot(kind="bar")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
# Task A
df["department"].value_counts().plot(kind="bar")
plt.title('Кол-во сотрудников по каждому отделу')
plt.ylabel("count")
plt.show()
```

```python
# Task B
df["age"].plot(kind="hist", bins=5)
plt.title("Распределение возрастов")
plt.xlabel("age")
plt.show()
```

```python
# Task C
df["salary"].plot(kind="box")
plt.title("Зарплатная свеча")
plt.ylabel("salary")
plt.show()
```

```python
# Task D
df.groupby("department")["salary"].mean().plot(kind="bar")
plt.title("Средняя зарпалата по отделам")
plt.ylabel("average salary")
plt.show()
```
