```python
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("../data/raw/titanic.csv")
df.head()
```

```python
features = ["Pclass", "Sex", "Age", "Fare", "Embarked"]
target = "Survived"

X = df[features]
y = df[target]

X[:10]
```

```python
numeric_features = ["Pclass", "Age", "Fare"]
categorical_features = ["Sex", "Embarked"]
```

```python
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])
```

```python
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore")),
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
logreg_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000))
])
```

```python
logreg_scores = cross_val_score(
    logreg_model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)
logreg_scores

print(f"LogReg mean accuracy: {logreg_scores.mean():.4f}")
print(f"LogReg std accuracy : {logreg_scores.std():.4f}")
```

```python
tree_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", DecisionTreeClassifier(random_state=42))
])
```

```python
tree_scores = cross_val_score(
    tree_model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

print(f"LogReg mean accuracy: {tree_scores.mean():.4f}")
print(f"LogReg std accuracy : {tree_scores.std():.4f}")
```

```python
rf_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42)),
])
```

```python
rf_scores = cross_val_score(
    rf_model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

print(f"LogReg mean accuracy: {rf_scores.mean():.4f}")
print(f"LogReg std accuracy : {rf_scores.std():.4f}")
```

```python
results = pd.DataFrame({
    "model": ["LogisticRegression", "DecisionTree", "RandomForest"],
    "mean_accuracy": [logreg_scores.mean(), tree_scores.mean(), rf_scores.mean()],
    "std_accuracy": [logreg_scores.std(), tree_scores.std(), rf_scores.std()]

})

results.sort_values("mean_accuracy", ascending=False)
```

```python
# Task A
features = ["Pclass", "Sex", "Age", "Fare", "Embarked"]
target = "Survived"

X = df[features]
y = df[target]

numeric_features = ["Pclass",  "Age", "Fare"]
categorical_features = ["Sex", "Embarked"]

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])
```

```python
# Task B
logreg_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000))
])

tree_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", DecisionTreeClassifier(random_state=42))
])
rf_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])
```

```python
# Task C
logreg_scores = cross_val_score(
    logreg_model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

tree_scores = cross_val_score(
    tree_model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)

rf_scores = cross_val_score(
    rf_model,
    X,
    y,
    cv=5,
    scoring="accuracy"
)
```

```python
# Task D
print(f"logreg mean - {logreg_scores.mean():.4f}\nlogreg std -{logreg_scores.std():.4f}")
print(f"decision tree mean - {tree_scores.mean():.4f}\ntree std - {tree_scores.std():.4f}")
print(f"random forest mean - {rf_scores.mean():.4f}\nrf std - {rf_scores.std():.4f}")
```

```python
# Task E
results = pd.DataFrame({
    "model": ["LogisticRegression", "DecisionTreeClassifier", "RandomForestClassifier"],
    "mean_accuracy": [logreg_scores.mean(), tree_scores.mean(), rf_scores.mean()],
    "std_accuracy": [logreg_scores.std(), tree_scores.std(), rf_scores.std()],
}
)

results.sort_values("mean_accuracy", ascending=False)
```

## Выводы

- Модели нужно сравнивать в одинаковых условиях.
- Cross-validation даёт более надёжную оценку, чем один train/test split.
- Mean accuracy показывает средний уровень качества.
- Std accuracy показывает стабильность модели между фолдами.
- Лучшая baseline-модель — не всегда та, у которой просто максимальная одна цифра.
