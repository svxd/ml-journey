```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "employee_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "Diana"],
    "department_id": [10, 20, 10, 30]
})

departments = pd.DataFrame({
    "department_id": [10, 20],
    "department_name": ["IT", "HR"]
})

employees
```

```python
employees.merge(departments, on="department_id", how="inner")
```

```python
employees.merge(departments, on="department_id", how="left")
```

```python
employees.merge(departments, on="department_id", how="outer")
```

```python
# Task dataset_01
employees = pd.DataFrame({
    "employee_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "Diana"],
    "department_id": [10, 20, 10, 30]
})

departments = pd.DataFrame({
    "department_id": [10, 20, 40],
    "department_name": ["IT", "HR", "Finance"]
})
employees
```

```python
departments
```

```python
# Task A
employees.merge(departments, on="department_id", how="left")
```

```python
# Task dataset_02
salaries = pd.DataFrame({
    "employee_id": [1, 2, 3],
    "salary": [50000, 70000, 90000]
})
salaries
```

```python
# Task B
employees.merge(salaries, on="employee_id", how="left")
```

```python
# Task C
clear_dataset = (
    employees
    .merge(departments, on="department_id", how="left")
    .merge(salaries, on="employee_id", how="left")
)
clear_dataset
```

```python
# Task D
clear_dataset.isna().sum()
```

NaN появились, потому что для некоторых ключей не нашлось совпадений
в связанных таблицах:
- у Diana department_id = 30, но такого id нет в departments
- у Diana employee_id = 4, но такого id нет в salaries
Поэтому после left merge соответствующие значения стали NaN.

Task E\
